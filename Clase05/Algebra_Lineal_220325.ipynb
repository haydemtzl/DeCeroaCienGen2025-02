```python
# Operaciones con matrices

A = np.array([[1,3,2],
             [1,0,0],
             [1,2,2]])
A
```

```python
# Multiplicación por escalar
A * 2
```

```python
B = np.array([[1,0,5],
              [7,5,0],
              [2,1,1]])
```

```python
# Suma de matrices
A + B
```

```python
# Resta de matrices
A - B
```

```python
# Para ver la dimensión de una matriz
A.shape
```

```python
# Para ver la cantidad de elementos
A.size
```

```python
# Multiplicación de matrices

A1 = np.arange(1,13).reshape(3,4)
A1
```

```python
B1 = np.arange(8).reshape(4,2)
B1
```

```python
A1 @ B1
```

```python
B1 @ A1
```

```python
# Matriz identidad
I = np.eye(2)
I
```

```python
A = np.array([[4,7],
              [2,6]])
A
```

```python
# Multiplicar matriz cuadrada por su matriz identidad
A @ I
```

```python
# Calculo del determinante
np.linalg.det(A)
```

```python
B = np.array([[4,7,1],
              [2,6,1],
              [1,1,1]])
np.linalg.det(B)
```

```python
# Matriz transpuesta

C = np.arange(6).reshape(3,2)
C
```

```python
np.transpose(C)
```

```python
#Calculando la matriz inversa

A_inv = np.linalg.inv(A)
A_inv
```

```python
D = np.array([[1,1,0],
              [2,-1,2],
              [0,3,0]])
```
