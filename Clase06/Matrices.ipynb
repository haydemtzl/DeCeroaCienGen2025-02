# Matriz identidad

```python
import numpy as np
```

```python
I = np.eye(2)
I
```

```python
A = np.array([[4,7],[2,6]])
A
```

```python
A @ I
```

# Determinante

```python
np.linalg.det(A)
```

```python
A1 = np.array([[1,1,0],[2,-1,2],[0,3,0]])
A1
```

```python
np.linalg.det(A1)
```

# Matriz transpuesta

```python
A2 = np.arange(6).reshape(3,2)
A2
```

```python
np.transpose(A2)
```

# Matriz inversa

```python
A_inv = np.linalg.inv(A)
A_inv
```

```python
A1_inv = np.linalg.inv(A1)
A1_inv
```
