```python
# Importar matplotlib

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Grafica simple
x = [1,2,3,4]
y = [10,15,7,12]

plt.plot(x,y) #Pidiendo que me grafique x, y 
plt.show() #Mostrar la grafica
```

```python
# Grafica con etiquetas
x = [1,2,3,4]
y = [10,15,7,12]

plt.plot(x,y) #Grafique
plt.title("Grafiquita")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
# Grafico de y = x^2

x = np.array([0,1,2,3,4,5])
y = x**2

plt.plot(x,y)
plt.title("Grafico sencillo de y = x^2")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

```python
# Grafico de y = x^2 con leyenda de linea

x = np.array([0,1,2,3,4,5])
y = x**2

plt.plot(x,y,label = "linea de x vs y")
plt.title("Grafico sencillo de y = x^2")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.legend()
plt.show()
```

```python
# Grafico de y = x^2 con leyenda de linea

x = np.array([0,1,2,3,4,5])
y = x**2

plt.plot(x,y,label = "linea de x vs y")
plt.title("Grafico sencillo de y = x^2")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.legend()
plt.grid(True) #Para la cuadricula
plt.show()
```

```python
# Grafica de barras

categorias = ["A","B","C","D"]
valores = [10,15,7,12]

plt.bar(categorias, valores, color="skyblue") # Aqui estoy utilizando "bar" que es para especificar que es una grfica de barras
plt.title("Grafica de barras")
plt.show
```

```python
# Grfica de pastel simple

etiquetas = ["Caballos","Cerdos","Perros","Vacas"]
porcentajes = [15,30,45,10]

plt.pie(porcentajes, labels = etiquetas, autopct='%1.1f%%', startangle = 90) # Le indico que es de pastel
plt.title("Grfico de pastel")
plt.show
```

```python
# Grfica de pastel con explode (separar un valor especifico)

etiquetas = ["Caballos","Cerdos","Perros","Vacas"]
porcentajes = [15,30,45,10]
explode = (0,0.1,0,0)

plt.pie(porcentajes, explode = explode, labels = etiquetas, autopct='%1.1f%%', startangle = 90) # Le indico que es de pastel y le doy el explode
plt.title("Grfico de pastel")
plt.show()
```

```python
# Grafico con colores estilos de linea y marcadores

x = [0,1,2,3,4]
y = [i+1 for i in x]

plt.plot(x,y, color="blue", linewidth=2, linestyle = "--", marker="o", markersize=8, markerfacecolor="red", markeredgecolor="black" )
plt.title("Lnea con estilo")
plt.grid(True)
plt.show()
```

```python
# ESTILOS DE LINEAS Y MARCADORES
# ancho de linea = linewidth
# estilo de linea = linestyle

x = np.linspace(0,5, 11)
y = x ** 2

plt.plot(x, x+1, color="red", linewidth=0.25)
plt.plot(x, x+2, color="red", linewidth=0.50)
plt.plot(x, x+3, color="red", linewidth=1.00)
plt.plot(x, x+4, color="red", linewidth=2.00)

# posibles opciones linestype -, , -., :, steps
plt.plot(x, x+5, color="green", lw=3, linestyle='-')
plt.plot(x, x+6, color="green", lw=3, ls='-.')
plt.plot(x, x+7, color="green", lw=3, ls=':')

# lineas parametrizadas
line, = plt.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # formato: longitud de linea, longitud de espacio, ...

# posibles simbolos del marcas: marker = '+', 'o', '*', 's', ',', '.',bb '1', '2', '3', '4', ...
plt.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
plt.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
plt.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
plt.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# tamao y color de la marca
plt.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
plt.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
plt.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
plt.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
# Grupos para diagrama de caja o Boxplot

grupo1 = np.random.normal(0,1,100)
grupo2 = np.random.normal(0,2,100)
grupo3 = np.random.normal(0,3,100)

plt.boxplot([grupo1, grupo2, grupo3], vert=True, patch_artist=True)
plt.title("Boxplot")
plt.show()
```

```python
# Histograma

from random import sample

datos = sample(range(1,1000), 100) # Le estoy pidiendo que me genere una lista de 100 valores en un rango del 1 al 1000

plt.hist(datos, bins=8, color="pink") #bins es el numero de divisiones de mi histograma
plt.title("Histograma")
plt.show()
```

```python
# Scatter Plot (grfico de dispersin)

x = [1,2,3,4,5]
y = [2,4,1,8,7]

plt.scatter(x,y, color="purple")
plt.title("Scatter Plot")
plt.show()
```

```python
#Para guardar grficas

# Matlplotlib me permite guardar las grficas en formatos PNG, JPG, EPS, SVG, PGF y PDF
# Para guardar una figura en un archivo, puedo utilizar el mtodo savefig de la clase Figure
# Antes de crear una grafica tengo definir una variable que almacene la figura

fig = plt.figure()
plt.scatter(x,y, color = "purple")
plt.title("Scatter Plot Simple")

fig.savefig("figura.png")
```

```python
# GUARDANDO Grafica de pastel simple

etiquetas = ["Caballos","Cerdos","Perros","Vacas"]
porcentajes = [15,30,45,10]

fig = plt.figure() #Se crea variabe para guardar la figura
plt.pie(porcentajes, labels=etiquetas, shadow = True, autopct='%1.1f%%', startangle = 90) # Le indico que es de pastel
plt.title("Grfico de pastel")
fig.savefig("pastel.png") # La funcin que me guarda la figura tal cual
```
