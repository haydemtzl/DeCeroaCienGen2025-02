```python
# Importar las librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar mis datos

data = pd.read_csv("2titanic.csv")
```

```python
# Leer los primeros 5 renglones de mi dataset

data.head()
```

```python
# Para ver el tamaño de mi dataset, para ver cuántos renglones y columnas tengo
data.shape
```

```python
#La función count me permite saber cuantos elementos hay en cada columna
data.count()
```

```python
col_names = data.columns.tolist() # Convierte los nombres de las columnas en una lista y las variable col_names
col_names
```

```python
for column in col_names:
    print("Valores nulos en", column, ":", data[column].isnull().sum()) # esta función hace una sumatoria de todos los valores nulos
                                                                        # de cada columna
```

```python
# Vamos a revisar la estadística descriptiva de mi dataset
data.describe()
```

```python
def sex_transform(x):
    d = {'male':'M','female':'F'} # Esta función va a recibir un diccionario y un valor y va a retornar el valor.
    return(d[x])

data['Sex'] = data['Sex'].apply(sex_transform)
```

```python
data.head()
```

```python
# Copiar la columna sex a una nueva columna que se llame Gender

data['Gender'] = data['Sex']
data.head()
```

# Exploración de datos

```python
# Cuántas y cuáles pasajeros se subieron sin pagar?

data[data.Fare == 0]
```

```python
# Tabla cruzada entre supervivencia y sexo
pd.crosstab(data.Survived, data.Sex)
```

```python
# Agrupar los datos por clase y sexo, después solo mostrar la suma de los que sobrevivieron
pclass_sex_survival_count = data.groupby(['Pclass','Sex'])['Survived'].sum()
pclass_sex_survival_count
```

```python
# Tabla cruzada entre PClass y Sexo
pd.crosstab(data.Pclass, data.Sex)
```

```python
#Gráfica de la columna survived
data.Survived.value_counts().plot(kind = 'bar')
plt.title("Sobrevivieron / cuenta total")
```

```python
#Grafica de la columna survided por porcentaje
data.Survived.value_counts(normalize = True).plot(kind = "bar", alpha = 0.4)
plt.title('Sobrevivieron / porcentaje total')
```

```python
# Grafica de sexo y survived en cuenta total
data.Sex[data.Survived == 1].value_counts().plot(kind = 'barh', color = "pink", alpha = 0.7)
plt.title("Sobrevivieron totales - M vs F")
```

```python
#Grafica de sexo y survived en porcentajes
data.Sex[data.Survived == 1].value_counts(normalize = True).plot(kind = "barh", color = "pink", alpha = 0.9)
plt.title("Sobrevivieron porcentajes M vs F")
```

```python
# Grafica survived por clase
data.Pclass[data.Survived == 1].value_counts().plot(kind = "bar", color = "purple")
plt.title("Sobrevivientes por clase - Totales")
```

```python
# Grafica survived por clase en porcentajes
data.Pclass[data.Survived == 1].value_counts(normalize = True).plot(kind = "bar", color = "purple", alpha = 0.7)
plt.title("Sobrevivientes por clase - Porcentajes")
```

```python
# Grafica por clase en porcentajes - pastel
data.Pclass[data.Survived == 1].value_counts(normalize = True).plot(kind = "pie")
plt.title("Sobrevivientes por clase - Porcentajes")
```

```python
# Nueva gráfica para ver si hay una relación entre tipo de ticket y edad

fig = plt.figure(figsize = (20,10))
for clase in [1,2,3]:
    data.Age[data.Pclass == clase].plot(kind = "kde")
plt.legend(("1ra Clase","2da Clase","3era Clase"))
```
